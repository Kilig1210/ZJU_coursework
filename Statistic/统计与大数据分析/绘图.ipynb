```python
import matplotlib.pyplot as plt

# 提供的数据
years = [2018, 2019, 2020, 2021, 2022]
market_size = [7.01, 8.62, 10.12, 12.52, 14.33]
growth_rate = [0, 22.97, 17.40, 23.72, 14.46]

plt.rcParams['font.sans-serif']=['SimHei'] #用来正常显示中文标签
plt.rcParams['axes.unicode_minus']=False #用来正常显示负号

# 创建一个图形对象和轴对象
fig, ax1 = plt.subplots()

# 绘制柱形图，减少宽度
ax1.bar(years, market_size, color='#1f77b4', alpha=0.7, label='市场规模', edgecolor='black', linewidth=1.5, width=0.6)

# 创建第二个轴对象，共享x轴
ax2 = ax1.twinx()

# 绘制折线图
ax2.plot(years, growth_rate, color='#ff7f0e', marker='o', label='增长率', linewidth=2)

# 设置标签和标题
ax1.set_xlabel('年份')
ax1.set_ylabel('市场规模', color='#1f77b4')
ax2.set_ylabel('增长率 (%)', color='#ff7f0e')
plt.title('市场规模和增长率趋势')

# 显示图例
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

# 设置网格线
ax1.grid(True, linestyle='--', alpha=0.5)

# 显示图形
plt.show()


```

```python
import matplotlib.pyplot as plt

# 提供的数据
years = [2018, 2019, 2020, 2021, 2022]
market_size = [7.01, 8.62, 10.12, 12.52, 14.33]
growth_rate = [ 0, 22.97, 17.40, 23.72, 14.46]

plt.rcParams['font.sans-serif']=['SimHei'] #用来正常显示中文标签
plt.rcParams['axes.unicode_minus']=False #用来正常显示负号

# 创建一个图形对象和轴对象
fig, ax1 = plt.subplots(figsize=(8, 6))  # 调整图形大小

# 绘制柱形图，减少宽度
ax1.bar(years, market_size, color='#1f77b4', alpha=0.7, label='市场规模', edgecolor='black', linewidth=1.5, width=0.6)

# 创建第二个轴对象，共享x轴
ax2 = ax1.twinx()

# 绘制折线图
ax2.plot(years, growth_rate, color='#ff7f0e', marker='o', label='增长率', linewidth=2)

# 设置标签和标题，调整字体
ax1.set_xlabel('年份', fontsize=12)
ax1.set_ylabel('市场规模', fontsize=12)
ax2.set_ylabel('增长率 (%)', fontsize=12)
plt.title('市场规模和增长率趋势', fontsize=14)

# 显示图例，调整图例字体
ax1.legend(fontsize=10)
ax2.legend(fontsize=10)

# 设置网格线
ax1.grid(True, linestyle='--', alpha=0.5)

# 调整刻度字体大小
ax1.tick_params(axis='both', labelsize=10)
ax2.tick_params(axis='y', labelsize=10)

# 显示图形
plt.tight_layout()  # 调整布局，防止标签重叠
plt.show()
```
